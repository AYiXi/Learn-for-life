```python
class Base():
    def hello(self):
        print('hello')

class A(Base):
    # 2
    def hello(self):
        return super(A, self).hello()
    
class B(Base):
    # 3
    def hello(self):
        return super().hello()

A().hello()
B().hello()              
```

```python
# 防止参数顺序搞混
def add(a, b, *, c):
    return a + b + c

print(add(1, 2, c=3))
```

```python
def fun1(*args):
    for i, j in enumerate(args):
        print(i, j)

def fun2(**kwargs):
    for k, v in kwargs.items():
        print(k,'=', v)        
    
fun1('a', 'b', 'c')
fun1(*['a', 'b', 'c'])
fun2(a=1, b=2, c=3)
fun2(**dict(a=1, b=2, c=3))
```
