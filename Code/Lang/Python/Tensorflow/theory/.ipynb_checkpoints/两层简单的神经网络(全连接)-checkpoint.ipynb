# 反向传播

```python
import numpy as np
```

```python
# 一次喂入多少数据,数据不宜过大
BATCH_SIZE=8
seed=23455

# 基于seed产生随机数
rng = np.random.RandomState(seed)

# 随机数返回32X2的矩阵 表示32组体积和重量  作为输入数据集
X = rng.rand(32, 2)

# 从这32X2的矩阵中 取出一行 判断如果和小于1 给Y赋值1 否则赋值0
# 作为输入数据集的标签(正确答案)
Y = [[int(x0+x1<1)] for (x0,x1) in X]
print('\nX:\n', X)
print('\nY:\n', Y)

# 定义神经网络的输入，参数和输出，定义前向传播过程
x = tf.placeholder(tf.float32, shape=(None,2))
y_true = tf.placeholder(tf.float32, shape=(None,1))

w1 = tf.Variable(tf.random_normal(shape=[2,3], stddev=1, seed=1))
w2 = tf.Variable(tf.random_normal(shape=[3,1], stddev=1, seed=1))

a = tf.matmul(x, w1)
y_predict = tf.matmul(a, w2)

# 定义损失函数和反向传播算法
loss = tf.reduce_mean(tf.square(y_predict-y_true))
train_step = tf.train.GradientDescentOptimizer(0.001).minimize(loss)
# train_step = tf.train.AdamOptimizer(0.001).minimize(loss)
# train_step = tf.train.MomentumOptimizer(0.001).minimize(loss)

with tf.Session() as sess:
    init_op = tf.global_variables_initializer()
    sess.run(init_op)
    
    # 输出未经训练的参数取值
    print('\nw1:\n', sess.run(w1))
    print('\nw2:\n', sess.run(w2))
    
    STEPS=3000
    print('\n')
    for i in range(STEPS):
        start = (i*BATCH_SIZE) % 32
        end = start + BATCH_SIZE
        sess.run(train_step, feed_dict={x:X[start:end], y_true:Y[start:end]})
        if i % 100 == 0:
            total_loss = sess.run(loss, feed_dict={x:X, y_true:Y})
            print('After %d training steps, \tloss on all data is %g' % (i, total_loss))
    
    # 输出训练结束后参数取值
    print('\nw1:\n', sess.run(w1))
    print('\nw2:\n', sess.run(w2))    
```

# 损失函数

```python
# 一次喂入多少数据,数据不宜过大
BATCH_SIZE=8
seed=23455

# 基于seed产生随机数
rng = np.random.RandomState(seed)

# 随机数返回32X2的矩阵 表示32组体积和重量  作为输入数据集
X = rng.rand(32, 2)

# 从这32X2的矩阵中 取出一行 判断如果和小于1 给Y赋值1 否则赋值0
# 作为输入数据集的标签(正确答案)
Y = [[x1+x2+(rng.rand()/10.0-0.05)] for (x1,x2) in X]

# 定义神经网络的输入，参数和输出，定义前向传播过程
x = tf.placeholder(tf.float32, shape=(None,2))
y_true = tf.placeholder(tf.float32, shape=(None,1))

w1 = tf.Variable(tf.random_normal(shape=[2,1], stddev=1, seed=1))

y_predict = tf.matmul(x, w1)

# 定义损失函数和反向传播算法
loss = tf.reduce_mean(tf.square(y_predict-y_true))
train_step = tf.train.GradientDescentOptimizer(0.001).minimize(loss)
# train_step = tf.train.AdamOptimizer(0.001).minimize(loss)
# train_step = tf.train.MomentumOptimizer(0.001).minimize(loss)

with tf.Session() as sess:
    init_op = tf.global_variables_initializer()
    sess.run(init_op)
    
    # 输出未经训练的参数取值
    print('\nw1:\n', sess.run(w1))
    print('\nw2:\n', sess.run(w2))
    
    STEPS=20000
    print('\n')
    for i in range(STEPS):
        start = (i*BATCH_SIZE) % 32
        end = start + BATCH_SIZE
        sess.run(train_step, feed_dict={x:X[start:end], y_true:Y[start:end]})
        if i % 500 == 0:
            total_loss = sess.run(loss, feed_dict={x:X, y_true:Y})
            print('After {} training steps,w1 is {}'.format(i, sess.run(w1)))
    
    # 输出训练结束后参数取值
    print('\nw1:\n', sess.run(w1))
```

## 自定义损失函数

```python
# 一次喂入多少数据,数据不宜过大
BATCH_SIZE=8
seed=23455
COST=1
PROFIT=9

# 基于seed产生随机数
rng = np.random.RandomState(seed)

# 随机数返回32X2的矩阵 表示32组体积和重量  作为输入数据集
X = rng.rand(32, 2)

# 从这32X2的矩阵中 取出一行 判断如果和小于1 给Y赋值1 否则赋值0
# 作为输入数据集的标签(正确答案)
Y = [[x1+x2+(rng.rand()/10.0-0.05)] for (x1,x2) in X]

# 定义神经网络的输入，参数和输出，定义前向传播过程
x = tf.placeholder(tf.float32, shape=(None,2))
y_true = tf.placeholder(tf.float32, shape=(None,1))

w1 = tf.Variable(tf.random_normal(shape=[2,1], stddev=1, seed=1))

y_predict = tf.matmul(x, w1)

# 定义损失函数和反向传播算法
# 自定义损失函数使得预测少了损失达，于是模型应该偏向多的方向预测
loss = tf.reduce_mean(tf.where(tf.greater(y_predict, y_true), 
                              (y_predict-y_true)*COST,
                              (y_true-y_predict)*PROFIT,))
train_step = tf.train.GradientDescentOptimizer(0.001).minimize(loss)
# train_step = tf.train.AdamOptimizer(0.001).minimize(loss)
# train_step = tf.train.MomentumOptimizer(0.001).minimize(loss)

with tf.Session() as sess:
    init_op = tf.global_variables_initializer()
    sess.run(init_op)
    
    # 输出未经训练的参数取值
    print('\nw1:\n', sess.run(w1))
    print('\nw2:\n', sess.run(w2))
    
    STEPS=20000
    print('\n')
    for i in range(STEPS):
        start = (i*BATCH_SIZE) % 32
        end = start + BATCH_SIZE
        sess.run(train_step, feed_dict={x:X[start:end], y_true:Y[start:end]})
        if i % 500 == 0:
            total_loss = sess.run(loss, feed_dict={x:X, y_true:Y})
            print('After {} training steps,w1 is {}'.format(i, sess.run(w1)))
    
    # 输出训练结束后参数取值
    print('\nw1:\n', sess.run(w1))
```

# 学习率

```python
w = tf.Variable(tf.constant(10, dtype=tf.float32))

loss = tf.square(w+1)

train_step=tf.train.GradientDescentOptimizer(0.2).minimize(loss)

with tf.Session() as sess:
    init_op = tf.global_variables_initializer()
    sess.run(init_op)
    print(f' w is {sess.run(w)}, loss is {sess.run(loss)}')    
    for i in range(40):
        sess.run(train_step)
        w_val = sess.run(w)
        loss_val = sess.run(loss)
        print(f'After {i} steps: w is {w_val}, loss is {loss_val}')
```

## 动态衰减学习率

```python
LEARN_RATE_BASE = 0.1 #最初学习率
LEARN_RATE_DECAY = 0.99 #学习率衰减率
LEARN_RATE_STEP = 1 #喂入多少轮BATCH_SIZE后，更新一次学习率，一般设为：样本总数/BATCH_SIZE。

# 运行了几轮的BATCH_SIZE计数器，初值为0，设为不做训练
global_step = tf.Variable(0, trainable=False)

# 定义指数下降学习率
learning_rate = tf.train.exponential_decay(LEARN_RATE_BASE,
                                           global_step,
                                           LEARN_RATE_STEP,
                                           LEARN_RATE_DECAY,
                                           staircase=True)

w = tf.Variable(tf.constant(10, dtype=tf.float32))

loss = tf.square(w+1)

train_step=tf.train.GradientDescentOptimizer(learning_rate).minimize(loss)

with tf.Session() as sess:
    init_op = tf.global_variables_initializer()
    sess.run(init_op)
    for i in range(40):
        sess.run(train_step)
        w_val = sess.run(w)
        loss_val = sess.run(loss)
        print(f'After {i} steps: w is {w_val}, loss is {loss_val}')
```

# 滑动平均

```python
# 定义变量以及滑动平均类
# 定义一个32位浮点变量，初始值0.0，代码不断更新优化w1，滑动平均做了w1的影子
w1 = tf.Variable(0, dtype=tf.float32)

# 运行了几轮的计数器，初值为0，设为不做训练
global_step = tf.Variable(0, trainable=False)

# 实例化滑动平均类，给删减率0.99，当前轮数global_step
MOVING_AVERAGE_DECAY = 0.99
ema = tf.train.ExponentialMovingAverage(MOVING_AVERAGE_DECAY, global_step)

# ema.apply后的括号里使更新列表，每次运行sess.run(ema_op)时，对更新列表中的元素求滑动平均值。
# 在实际应用中会使用tf.trainable_variables()自动将所有待训练的参数汇总为列表
# ema_op = ema.apply([w1])
ema_op = ema.apply(tf.trainable_variables())

with tf.Session() as sess:
    init_op = tf.global_variables_initializer()
    sess.run(init_op)
    # 用ema.average(w1)获取w1的滑动平均值(要运行多个节点，作为列表中的元素列出，写在sess.run()中)
    # 打印出当前的w1和w1滑动平均值
    print(sess.run([w1, ema.average(w1)]))
    
    # 参数w1的赋值为1
    sess.run(tf.assign(w1, 1))
    sess.run(ema_op)    
    print(sess.run([w1, ema.average(w1)]))
    
    # 更新step和w1的值，模拟100轮后，参数w1变成10
    sess.run(tf.assign(global_step, 100))
    sess.run(tf.assign(w1, 10))  
    sess.run(ema_op)
    print(sess.run([w1, ema.average(w1)]))

    # 每次sess.run会更新一次w1的滑动平均值
    sess.run(ema_op)
    print(sess.run([w1, ema.average(w1)]))
    
    sess.run(ema_op)
    print(sess.run([w1, ema.average(w1)]))
    
    sess.run(ema_op)
    print(sess.run([w1, ema.average(w1)]))
    
    sess.run(ema_op)
    print(sess.run([w1, ema.average(w1)]))
    
    sess.run(ema_op)
    print(sess.run([w1, ema.average(w1)]))
    
    sess.run(ema_op)
    print(sess.run([w1, ema.average(w1)]))
# 更改MOVING_AVERAGE_DECAY为0.1，观察影子追随速度
```

# 正则化减少过拟合

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')

BATCH_SIZE = 30
seed = 2

# 基于seed产生随机数
rdm = np.random.RandomState(seed)

# 随机数返回300X2的矩阵，表示300组坐标点(x0,x1)作为输入数据集
X = rdm.randn(300, 2)

# 如果和的平方小于2，赋值为1，否则赋值为0，之内的颜色为red，否则为blue，便于区分
Y_true = [int(x0*x0 + x1*x1 < 2) for (x0, x1) in X]
Y_color = [['red' if y else 'blue'] for y in Y_true]

# 对数据集X和标签Y进行shape整理
X = np.vstack(X).reshape(-1, 2)
Y_true = np.vstack(Y_true).reshape(-1, 1)

print(X)
print(Y_true)
print(Y_color)

plt.scatter(X[:, 0], X[:, 1], c=np.squeeze(Y_color))
plt.show()

# 定义神经网络的输入，参数和输出，定义前向传播过程
def get_weight(shape, regularizer):
    w = tf.Variable(tf.random_normal(shape=shape), dtype=tf.float32)
    # 把每一个w的正则化损失加入到总损失losses中
    tf.add_to_collection(name='losses', value=tf.contrib.layers.l2_regularizer(regularizer)(w))
    return w

def get_bias(shape):
    b = tf.Variable(tf.constant(0.01, shape=shape))
    return b

x = tf.placeholder(tf.float32, shape=(None, 2))
y_true = tf.placeholder(tf.float32, shape=(None, 1))

w1 = get_weight([2, 11], 0.01)
b1 = get_bias([11])
y1 = tf.nn.relu(tf.matmul(x, w1) + b1)

w2 = get_weight([11, 1], 0.01)
b2 = get_bias([1])
y_predict =tf.matmul(y1, w2) + b2 # 输出层不过激活函数

# 定义损失函数
loss_mse = tf.reduce_mean(tf.square(y_predict-y_true))
# 均方误差的损失函数加上每一个正则化w的损失
loss_total = loss_mse + tf.add_n(tf.get_collection('losses'))

# 定义反向传播算法：不含正则化
# train_step = tf.train.GradientDescentOptimizer(0.0001).minimize(loss_mse)
train_step = tf.train.AdamOptimizer(0.0001).minimize(loss_mse)
                                                                
with tf.Session() as sess:
    init_op = tf.global_variables_initializer()
    sess.run(init_op)
                                                                
    for i in range(40000):
        start = (i*BATCH_SIZE) % 300
        end = start + BATCH_SIZE
        sess.run(train_step, feed_dict={
            x:X[start:end],
            y_true:Y_true[start:end],
        })
        if i % 2000 == 0:
            loss_mse_v = sess.run(loss_mse, feed_dict={
            x:X[start:end],
            y_true:Y_true[start:end],
        })
            print(f'after {i} steps, loss is {loss_mse_v}')
    
    # 生成二维网络坐标点
    xx, yy = np.mgrid[-3:3:0.01, -3:3:0.01]
    # 将xx，yy拉直，合并成一个2列的矩阵，得到一个网络坐标点的集合
    grid = np.c_[xx.ravel(), yy.ravel()]
    # 将坐标点喂入神经网络，probs为输出
    probs = sess.run(y_predict, feed_dict={x:grid})
    probs = probs.reshape(xx.shape)
    print('\nw1, b1, w2:', sess.run(w1), sess.run(b1), sess.run(w2))

plt.scatter(X[:, 0], X[:, 1], c=np.squeeze(Y_color))
# 给probs值为0.5的上色
plt.contour(xx, yy, probs, levels=[0.5])
plt.show()


# 定义反向传播算法：包含正则化
train_step = tf.train.AdamOptimizer(0.0001).minimize(loss_total)
                                                                
with tf.Session() as sess:
    init_op = tf.global_variables_initializer()
    sess.run(init_op)
                                                                
    for i in range(40000):
        start = (i*BATCH_SIZE) % 300
        end = start + BATCH_SIZE
        sess.run(train_step, feed_dict={
            x:X[start:end],
            y_true:Y_true[start:end],
        })
        if i % 2000 == 0:
            loss_mse_v = sess.run(loss_mse, feed_dict={
            x:X[start:end],
            y_true:Y_true[start:end],
        })
            print(f'after {i} steps, loss is {loss_mse_v}')
    
    # 生成二维网络坐标点
    xx, yy = np.mgrid[-3:3:0.01, -3:3:0.01]
    # 将xx，yy拉直，合并成一个2列的矩阵，得到一个网络坐标点的集合
    grid = np.c_[xx.ravel(), yy.ravel()]
    # 将坐标点喂入神经网络，probs为输出
    probs = sess.run(y_predict, feed_dict={x:grid})
    probs = probs.reshape(xx.shape)
    print('\nw1, b1, w2:', sess.run(w1), sess.run(b1), sess.run(w2))

plt.scatter(X[:, 0], X[:, 1], c=np.squeeze(Y_color))
# 给probs值为0.5的上色
plt.contour(xx, yy, probs, levels=[0.5])
plt.show()
```

# 神经网络搭建

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

class NN:
    BATCH_SIZE =30
    seed = 2
    STEPS = 40000
    LR_BASE = 0.001
    LR_DECAY = 0.999
    REGULARIZER = 0.01
    
    def generateds(self):
        rdm = np.random.RandomState(self.seed)
        X = rdm.randn(300,2)

        Y_true = [int (x0*x0 + x1*x1 <2) for (x0,x1) in X]

        Y_color = [['red' if y else 'blue'] for y in Y_true]

        X = np.vstack(X).reshape(-1,2)
        Y_true = np.vstack(Y_true).reshape(-1,1)
        return X,Y_true,Y_color
    
    # 定义神经网络的输入，参数和输出，定义前向传播过程
    def get_weight(self, shape, regularizer):
        W = tf.Variable(tf.random_normal(shape),dtype=tf.float32)
        tf.add_to_collection('losses',tf.contrib.layers.l2_regularizer(regularizer)(W))
        return W

    def get_bias(self, shape):
        b = tf.Variable(tf.constant(0.01,shape=shape))
        return b

    def forward(self, x, regularizer):
        W1 = self.get_weight([2,11], regularizer)
        b1 = self.get_bias([11])
        y1 = tf.nn.relu(tf.matmul(x,W1) + b1)

        W2 = self.get_weight([11,1], regularizer)
        b2 = self.get_bias([1])
        y = tf.matmul(y1,W2) + b2 # 输出层不过激活函数
        return y

    def backward(self):
        x = tf.placeholder(tf.float32,shape=(None,2))
        y_true = tf.placeholder(tf.float32,shape=(None,1))
        X,Y_true,Y_color = self.generateds()

        y_predict = self.forward(x,self.REGULARIZER)

        global_step = tf.Variable(0,trainable = False)

        lr = tf.train.exponential_decay(
            self.LR_BASE,
            global_step,
            300/self.BATCH_SIZE,
            self.LR_DECAY,
            staircase=True)

        loss_mse = tf.reduce_mean(tf.square(y_predict - y_true))
        loss_total = loss_mse + tf.add_n(tf.get_collection('losses'))

        # 有正则化
        train_step = tf.train.AdamOptimizer(lr).minimize(loss_total)

        with tf.Session() as sess:
            init_op = tf.global_variables_initializer()
            sess.run(init_op)
            for i in range(STEPS):
                start = (i * BATCH_SIZE) %300
                end = start + BATCH_SIZE
                sess.run(train_step,feed_dict={x:X[start:end],y_true:Y_true[start:end]})
                if i % 2000 == 0:
                    loss_mse_v = sess.run(loss_mse,feed_dict={x:X,y_true:Y_true})
                    print('Step:%d , loss：%f ' % (i,loss_mse_v))

            xx,yy = np.mgrid[-3:3:0.1,-3:3:0.1]
            grid = np.c_[xx.ravel(),yy.ravel()]
            probs = sess.run(y_predict,feed_dict={x:grid})
            probs = probs.reshape(xx.shape)
        # print "W1:\n",sess.run(W1)
        # print "b1:\n",sess.run(b1)
        # print "W2:\n",sess.run(W2)
        # print "b2:\n",sess.run(b2)

        plt.scatter(X[:,0],X[:,1],c=np.squeeze(Y_color))
        plt.contour(xx,yy,probs,levels=[.5])
        plt.show()

if __name__ == '__main__':
    nn = NN()
    nn.backward()
```

```python
import tensorflow as tf
import os
import time
from tensorflow.examples.tutorials.mnist import input_data

mnist = input_data.read_data_sets(r'F:\Jupyter notebook\Meachine Learning\data\mnist\input_data', one_hot=True)

# print(mnist.train.labels[0], mnist.train.images[0])


class Mnist:
    INPUT_NODE = 784  # 28X28
    OUTPUT_NODE = 10
    LAYEAR1_NODE = 500  # 隐藏层的节点个数
    BATCH_SIZE = 200
    STEPS = 50000
    MOVING_AVERAGE_DECAY = 0.99
    MODEL_SAVE_PATH = 'F:\Jupyter notebook\Meachine Learning\data\model'
    MODEL_NAME = 'mnist_model'
    LEARN_RATE_BASE = 0.1
    LEARN_RATE_DECAY = 0.99
    REGULARIZER = 0.0001

    # 定义神经网络的输入，参数和输出，定义前向传播过程
    def get_weight(self, shape, regularizer):
        w = tf.Variable(tf.truncated_normal(shape, stddev=0.1), dtype=tf.float32)
        if regularizer is not None:
            tf.add_to_collection('losses', tf.contrib.layers.l2_regularizer(regularizer)(w))
        return w

    def get_bias(self, shape):
        b = tf.Variable(tf.zeros(shape))
        return b

    def forward(self, x, regularizer):
        w1 = self.get_weight([self.INPUT_NODE, self.LAYEAR1_NODE], regularizer)
        b1 = self.get_bias([self.LAYEAR1_NODE])
        y1 = tf.nn.relu(tf.matmul(x, w1) + b1)

        w2 = self.get_weight([self.LAYEAR1_NODE, self.OUTPUT_NODE], regularizer)
        b2 = self.get_bias([self.OUTPUT_NODE])
        y = tf.matmul(y1, w2) + b2
        return y

    def backward(self, mnist_train):
        x = tf.placeholder(tf.float32, shape=(None, self.INPUT_NODE))
        y_true = tf.placeholder(tf.float32, shape=(None, self.OUTPUT_NODE))
        y_predict = self.forward(x, self.REGULARIZER)

        global_step = tf.Variable(0, trainable=False)

        ce = tf.nn.sparse_softmax_cross_entropy_with_logits(
            logits=y_predict,
            labels=tf.argmax(y_true, 1)
        )
        cem = tf.reduce_mean(ce)
        loss = cem + tf.add_n(tf.get_collection('losses'))

        learning_rate = tf.train.exponential_decay(
            self.LEARN_RATE_BASE,
            global_step,
            mnist.train.num_examples / self.BATCH_SIZE,
            self.LEARN_RATE_DECAY,
            staircase=True
        )

        # 有正则化
        train_step = tf.train.GradientDescentOptimizer(learning_rate).minimize(
            loss,
            global_step=global_step
        )

        ema = tf.train.ExponentialMovingAverage(self.MOVING_AVERAGE_DECAY, global_step)
        ema_op = ema.apply(tf.trainable_variables())
        with tf.control_dependencies([train_step, ema_op]):
            train_op = tf.no_op(name='train')

        saver = tf.train.Saver()

        with tf.Session() as sess:
            init_op = tf.global_variables_initializer()
            sess.run(init_op)
            
            # 回复w和b的值
            ckpt = tf.train.get_checkpoint_state(MODEL_SAVE_PATH)
            if ckpt and ckpt.model_checkpoint_path:
                saver.restore(sess, ckpt.model_checkpoint_path)

            for i in range(self.STEPS):
                xs, ys = mnist_train.train.next_batch(self.BATCH_SIZE)
                _, loss_value, step = sess.run([train_op, loss, global_step],
                                               feed_dict={x: xs, y_true: ys})
                if i % 1000 == 0:
                    print(f'Step: {i}, loss: {loss_value}')
                    saver.save(sess,
                               os.path.join(self.MODEL_SAVE_PATH, self.MODEL_NAME),
                               global_step=global_step
                               )

    def test(self, mnist_test):
        with tf.Graph().as_default() as g:
            x = tf.placeholder(tf.float32, shape=(None, self.INPUT_NODE))
            y_true = tf.placeholder(tf.float32, shape=(None, self.OUTPUT_NODE))
            y_predict = self.forward(x, self.REGULARIZER)
            
            
            
            # 所有参数加载时会赋值为各自的滑动平均值
            ema = tf.train.ExponentialMovingAverage(self.MOVING_AVERAGE_DECAY)
            ema_restore = ema.variables_to_restore()
            saver = tf.train.Saver(ema_restore)

            correct_prediction = tf.equal(tf.argmax(y_predict, 1), tf.argmax(y_true, 1))
            accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

            while True:
                with tf.Session() as sess:
                    ckpt = tf.train.get_checkpoint_state(self.MODEL_SAVE_PATH)
                    if ckpt and ckpt.model_checkpoint_path:
                        saver.restore(sess, ckpt.model_checkpoint_path)
                        global_step = ckpt.model_checkpoint_path.split('/')[-1].split('-')[-1]
                        accuracy_score = sess.run(accuracy,
                                                  feed_dict={x: mnist_test.test.images,
                                                             y_true: mnist_test.test.labels})
                        print(f'after {global_step} steps, accuracy is {accuracy_score}')
                    else:
                        print('no checkpoint file found!')
                time.sleep(5)


def train():
    Mnist().backward(mnist)


def test():
    Mnist().test(mnist)


if __name__ == '__main__':
    # train()
    test()
```
